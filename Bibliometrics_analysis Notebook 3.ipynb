DEFINING LABELS

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/content/3_table_1_with_abstracts.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Define the mapping for Tag1 labels
tag1_labels = {
    "1A": "Antiparasitic",
    "1B": "Antiviral",
    "1C": "Anti-vector",
    "1D": "Antiophidian",
    "1E": "Antifungal",
    "1F": "Antibacterial",
}

# Map the Tag1 codes to their descriptions
data['Tag1_Label'] = data['Tag1'].map(tag1_labels)

# Count the number of papers for each Tag1 label
paper_counts = data['Tag1_Label'].value_counts()

# Plot the results
plt.figure(figsize=(10, 6))
paper_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Anti-NTD activities of Natural products', fontsize=14)
plt.ylabel('Number of Papers', fontsize=14)
# Show the plot
plt.show()
```

```python
plt.figure(figsize=(5.5, 5.5))

sns.countplot(x='Tag5', data=papers, edgecolor='black')

plt.xlabel('Bioactivity class of Natural product', fontsize=14)
plt.ylabel('Number of Articles', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)

#plt.savefig('plot_bioactivity_class.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x='Year', y='Tag5', data=papers)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Bioactivity class', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size and DPI
sns.boxplot(x='Year', y='Tag5', data=papers[papers['Year'].between(2010, 2024)])
plt.xlabel('Year', fontsize=12)
plt.ylabel('Bioactivity class', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
